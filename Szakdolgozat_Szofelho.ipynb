```python
file_path = "/content/response.txt"

def remove_words_with_spaces(input_file, output_file):
    words_to_remove = ["még", "a", "az", "ez", "akkor", "hogy", "kell", "mert", "és"]
    punctuation = [".", ",", "!", "?", ";", ":", "'", "\"", "(", ")", "[", "]", "{", "}"]

    try:
        with open(input_file, 'r', encoding='latin-1') as file:
            content = file.read()

            # Szóképesítés a kereséshez
            for word in words_to_remove:
                search_term = f" {word} "
                content = content.replace(search_term, ' ')

            # Eltávolítás az írásjelek alapján
            for mark in punctuation:
                content = content.replace(mark, '')

        with open(output_file, 'w', encoding='latin-1') as output:
            output.write(content)

        print(f"A(z) {input_file} fájl tartalmát törölve és az eredményt elmentve a(z) {output_file} fájlba.")

    except FileNotFoundError:
        print(f"A(z) {input_file} fájl nem található.")

    except Exception as e:
        print(f"Hiba történt a fájl ({input_file}) feldolgozása során: {e}")

# Fájlok nevei
output_filename = "/content/modified_response2.txt"

# Függvény hívása
remove_words_with_spaces(file_path, output_filename)
```

```python
from wordcloud import WordCloud
import matplotlib.pyplot as plt

file_path = "/content/modified_response2.txt"

def create_wordcloud(file_path):
    try:
        with open(file_path, 'r', encoding='latin-1') as file:
            content = file.read()

            # Kizárt szavak
            excluded_words = ["lesz", "az", "úgy", "meg", "nem", "vagy", "ha"]

            # Kizárt szavak eltávolítása
            for word in excluded_words:
                content = content.replace(word, '')

            # WordCloud létrehozása
            wordcloud = WordCloud(width=800, height=400, background_color='white').generate(content)

            # Vizualizáció
            plt.figure(figsize=(10, 5))
            plt.imshow(wordcloud, interpolation='bilinear')
            plt.axis('off')
            plt.show()

    except FileNotFoundError:
        print(f"A(z) {file_path} fájl nem található.")

    except Exception as e:
        print(f"Hiba történt a fájl ({file_path}) feldolgozása során: {e}")

# Függvény hívása
create_wordcloud(file_path)
```
